```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
# resize
resized_img = cv2.resize(fix_img, (1000, 650))
plt.imshow(resized_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
reversed_0_img = cv2.flip(fix_img, 0) 
plt.imshow(reversed_0_img)
```

```python
reversed_1_img = cv2.flip(fix_img, 1) 
plt.imshow(reversed_1_img)
```

```python
cv2.imwrite('reversed1img.jpg',reversed_1_img)
```

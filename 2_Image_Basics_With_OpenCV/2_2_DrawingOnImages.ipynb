```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype = np.int16)
print(blank_img.shape)
plt.imshow(blank_img)
```

```python
# pt1 -> top left, pt2 -> bottom right
cv2.rectangle(blank_img, pt1=(284,100), pt2=(480,150), color=(0,255,0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(150,250), pt2=(275,380), color=(0,0,255), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(100,100), radius=50, color=(255,0,0), thickness=7)
plt.imshow(blank_img)
```

```python
# thickness = -1 -> fill in the circle
cv2.circle(blank_img, center=(400,400), color=(128,56,132), radius=55, thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(200,200), pt2=(400,300), color=(123,123,123), thickness=4)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(100,200,150), thickness = 10)
plt.imshow(blank_img)
```

```python
# text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text="Hello", org=(10,500), fontFace= font, fontScale= 3, color=(255,255,255), thickness= 4, lineType= cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text="Hello", org=(10,300), fontFace= font, fontScale= 3, color=(255,255,255), thickness= 4, lineType= cv2.LINE_4)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text="Hello", org=(10,400), fontFace= font, fontScale= 3, color=(255,255,255), thickness= 4, lineType= cv2.LINE_8)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype = np.int32)
vertices.shape
```

```python
pts = vertices.reshape(-1,1,2)
pts.shape
# 4,1,2 -> [
#            [
#               [,],
#               [,],
#               [,],
#               [,]
#            ] 
#          ]
# 1 sütun, 4 satır, her satır 2 eleman 
# (N,1,2)  "N" adet (1,2) boyutunda alt diziler oluştur
```

```python
cv2.polylines(blank_img, [pts], isClosed = True, color=(255,0,0), thickness =3)
plt.imshow(blank_img)
```

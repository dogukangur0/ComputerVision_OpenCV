```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/rainbow.jpg',0) # 0 -> grayscale 
plt.imshow(img, cmap="gray")
```

```python
_ , thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh, cmap="gray")
```

```python
_, thresh_inv = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh_inv, cmap="gray")
```

```python
_, thresh_trn = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh_trn, cmap="gray")
```

```python
_, thresh_tozero = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
plt.imshow(thresh_tozero, cmap="gray")
```

```python
_, thresh_tozero_inv = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh_tozero_inv, cmap="gray")
```

```python
img = cv2.imread('DATA/crossword.jpg',0)
plt.imshow(img, cmap="gray")
```

```python
def show_picture(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
show_picture(img)
```

```python
_, th1 = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY)
show_picture(th1)
```

```python
th2 = cv2.adaptiveThreshold(th1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 11, 8)
show_picture(th2)
```

```python
blended = cv2.addWeighted(src1 = th1, alpha = 0.6, src2 = th2, beta = 0.4, gamma = 0.4)
show_picture(blended)
```
